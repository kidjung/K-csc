### SPOOFING 사용 방법
+ 함수에 "ID, 시작, 끝, 어떤 아이디, 몇번째 필드, 어떻게"를 입력

```python
import pandas as pd
# 최대 줄 수 설정
pd.set_option('display.max_rows', 5000)
# 최대 열 수 설정
pd.set_option('display.max_columns', 5000)
```

```python
fname = "Pro_Cybersecurity_Car_Hacking_D_training-0.csv"
data = pd.read_csv(fname)
```

```python
# 데이터, 시작, 끝, 특정 ID, 몇번째를 무엇으로 바꿀껀지
def spoofing(df,start,end, ID , data_field=[],how=[] ):
    
    #시작 끝 설정
    total_count = len(df)
    start_index = int((total_count / 100) * start)
    end_index = int((total_count / 100) * end)

    # 초기 index
    count = start_index
    
    print(ID)
    while True:
        # 종료
        if count> end_index:
            break
        #spoofing 할 ID 검색
        if str(df["Arbitration_ID"][count]) == str(ID[0]):
            for i,num in enumerate(data_field):
                tem = "Data_"+str(num)
                df[tem][count] = how[i]
                df["Class"][count] ="Spoofing" 
        
        #하나씩 ECU 확인
        count += 1
        

    return df
```
