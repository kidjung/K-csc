### Fuzzing ATTACK 증강 설명
+ (데이터, 시작 시점, 끝나는 시점) 입력 받아서 랜덤 값 생성 
+ ID는 존재하는 ID중 랜덤하게 생성,  각 Datafield 들은 랜덤하게 16진수 2자리씩 생성
+ 기존 공격에서 DLC 가 모두 8, 증강도 DLC=8 인것만 생성

### 사용방법
+ (데이터, 시작 시점, 끝나는 지점)을 입력

```python
import pandas as pd
# 최대 줄 수 설정
pd.set_option('display.max_rows', 5000)
# 최대 열 수 설정
pd.set_option('display.max_columns', 5000)
import random
```

```python
fname = "Pro_Cybersecurity_Car_Hacking_D_training-0.csv"
data = pd.read_csv(fname)
```

```python
def fuzzing(df,start,end):
    total_count = len(df)
    start_index = int((total_count / 100) * start)
    end_index = int((total_count / 100) * end)
    adf = df
    # 초기 index
    count = start_index +2
    while True:
        if count > end_index:
            break
            
        #TIME 생성
        Time = adf["Timestamp"][count] + (adf["Timestamp"][count-2] - adf["Timestamp"][count-1])
            
        #랜덤 ID 생성
        ID_pool = list(set(list(adf["Arbitration_ID"].values))) # ID 중복 제거해서 리스트 생성
        tem = random.randrange(len(ID_pool))
        ran_ID = ID_pool[tem]
    
        #랜던 Datafield 생성
        ran_datafield =[]
        string_pool = "0123456789ABCDEF"
        for i in range(8):
            tem = random.choice(string_pool) + random.choice(string_pool)
            ran_datafield.append(tem)
        
        
        #형식 맞춰주기
        fuzzing = [Time]+ [ran_ID] +[8]+ ran_datafield + ["Fuzzing"]
        #fuzzing.append("Attack")
        
        #행에 추가해주기
        tem_s =pd.Series(fuzzing,index=['Timestamp','Arbitration_ID','DLC','Data_1',"Data_2","Data_3","Data_4","Data_5","Data_6","Data_7","Data_8","Class"])
        print(tem_s)
        front = adf[:count]
        back = adf[count:]
        front = front.append(tem_s,ignore_index = True)
        adf = pd.concat([front,back])
        
        #attack 간격 조정
        interval = random.randrange(1,8)
        count += interval
        
    return adf
```

```python
test = fuzzing(data,0,0.1)
```
